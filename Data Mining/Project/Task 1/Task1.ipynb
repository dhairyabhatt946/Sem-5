```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("OnlineRetail.csv", encoding='ISO-8859-1')
```

```python
df
```

```python
df.info()
```

```python
# necessary changes to solve the questions
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
# here the quantity columns of cancelled products contains negative values. so we need to filter our dataset to get correct answers of any calculations
df_clean = df[(df['Quantity'] > 0) & (df['CustomerID'].notnull())].copy()
```

```python
df_clean
```

```python
df_clean.info()
```

## 1. What is the total revenue generated (Quantity Ã— UnitPrice)?

```python
df_clean['TotalPrice'].sum()
```

## 2. Which product (StockCode) generated the highest total revenue?

```python
df_clean.groupby('StockCode')['TotalPrice'].sum().sort_values(ascending=False).head(1)
```

## 3. Which country (excluding UK) contributed the most to overall revenue?

```python
df_clean[df_clean['Country'] != 'United Kingdom'].groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(1)
```

## 4. Who are the top 5 customers based on total revenue?

```python
df_clean.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(5)
```

## 5. Find the top 10 most frequently purchased products and their total quantity sold.

```python
df_clean.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

## 6. How many unique invoices were generated in 2011?

```python
df_clean[df_clean['InvoiceDate'].dt.year == 2011]['InvoiceNo'].nunique()
```

## 7. What percentage of invoices were cancelled (start with 'C')?

```python
cancelled = df[df["InvoiceNo"].astype(str).str.startswith("C")]
(cancelled['InvoiceNo'].nunique() / df['InvoiceNo'].nunique()) * 100
```

## 8. Which day of the week sees the highest average revenue?

```python
df_clean['Weekday'] = df_clean['InvoiceDate'].dt.day_name()
df_clean.groupby('Weekday')['TotalPrice'].mean().sort_values(ascending=False).head(1)
```

## 9. What is the average number of unique items per invoice?

```python
df_clean.groupby('InvoiceNo')['StockCode'].nunique().mean()
```

## 10. Find the month with the highest number of unique customers.

```python
df_clean["Month"] = df_clean["InvoiceDate"].dt.to_period("M")
df_clean.groupby("Month")["CustomerID"].nunique().sort_values(ascending=False).head(1)
```

## 11. Calculate the mean, median, and standard deviation of UnitPrice.

```python
np.mean(df_clean["UnitPrice"]), np.median(df_clean["UnitPrice"]), np.std(df_clean["UnitPrice"])
```

## 12. Create a NumPy array of total prices and find how many transactions were above the average.

```python
total_prices = df_clean['TotalPrice'].values
np.sum(total_prices > total_prices.mean())
```

## 13. Normalize the Quantity column using Min-Max scaling.

```python
quantity = df_clean['Quantity'].values
normalized_qty = (quantity - quantity.min()) / (quantity.max() - quantity.min())
normalized_qty
```

## 14. How many unique customers are there using Numpy?

```python
np.unique(df_clean['CustomerID'].values).shape[0]
```

## 15. Create a correlation matrix between Quantity, UnitPrice, and TotalPrice.

```python
np.corrcoef(df_clean[["Quantity", "UnitPrice", "TotalPrice"]].values.T)
```

## 16. Use NumPy to identify invoices where the total number of items sold exceeds 100.

```python
items_per_invoice = df_clean.groupby('InvoiceNo')['Quantity'].sum()
np.sum(items_per_invoice.values > 100)
```

## 17. How many products were sold for a price greater than â‚¹10 and quantity above 5?

```python
filtered = df_clean[(df_clean["UnitPrice"] > 10) & (df_clean["Quantity"] > 5)]
filtered.shape[0]
```

## 18. Plot a bar chart showing total revenue generated by the top 5 countries.

```python
# total_revenue_top5_countries = df_clean.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(5)
# plt.bar(total_revenue_top5_countries.index, total_revenue_top5_countries.values)
# plt.xlabel("Country")
# plt.ylabel("Total Revenue")
# plt.title("Top 5 Countries by Revenue")
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

df_clean.groupby("Country")["TotalPrice"].sum().sort_values(ascending=False).head(5).plot(kind='bar', title="Top 5 Countries by Revenue")
plt.show()
```

## 19. Create a line chart of monthly total revenue across the dataset.

```python
# monthly_rev = df_clean.groupby(df_clean["InvoiceDate"].dt.to_period("M"))["TotalPrice"].sum()
# monthly_rev.plot(kind="line", marker='o', title="Monthly Revenue")
# plt.show()

monthly_rev = df_clean.groupby(df_clean["InvoiceDate"].dt.to_period("M"))["TotalPrice"].sum()
monthly_rev.index = monthly_rev.index.to_timestamp().strftime('%b %Y')
plt.plot(monthly_rev.index, monthly_rev.values)
plt.xticks(rotation=45)
plt.title("Monthly Revenue")
plt.show()
```

## 20. Create a pie chart of top 5 products based on quantity sold.

```python
top_products = df_clean.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='pie', autopct='%1.1f%%', title="Top 5 Products by Quantity")
plt.ylabel("")
plt.show()
```

## 21. Plot a histogram of UnitPrice.

```python
filtered_prices = df_clean[(df_clean["UnitPrice"] > 0) & (df_clean["UnitPrice"] < 100)]["UnitPrice"]
plt.hist(filtered_prices, bins=50)
plt.title("Unit Price Distribution (0â€“100)")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

## 22. Scatter plot between Quantity and UnitPrice to find outliers.

```python
df_clean.plot.scatter(x="Quantity", y="UnitPrice", title="Quantity vs UnitPrice")
plt.show()
```

## 23. Plot number of orders placed per weekday (Monday to Sunday).

```python
df_clean.groupby("Weekday")["InvoiceNo"].nunique().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
).plot(kind='bar', title="Orders per Weekday")
plt.show()
```

## 24. Create a boxplot to compare the distribution of UnitPrice among the top 5 selling products (by quantity).

```python
top5_products = df_clean.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(5).index
top5_data = df_clean[df_clean["Description"].isin(top5_products)]
top5_data.boxplot(column="UnitPrice", by="Description", grid=False)
plt.title("Unit Price Distribution for Top 5 Products")
plt.suptitle("")
plt.xticks(rotation=45)
plt.ylabel("Unit Price")
plt.show()
```

## 25. Plot a simple line graph showing monthly revenue for the top 3 countries.

```python
top3 = df_clean["Country"].value_counts().head(3).index
df_top3 = df_clean[df_clean["Country"].isin(top3)]
monthly_revenue = df_top3.groupby([df_top3["InvoiceDate"].dt.to_period("M"), "Country"])["TotalPrice"].sum().unstack()
monthly_revenue.index = monthly_revenue.index.to_timestamp().strftime('%b %Y')
monthly_revenue.plot()
plt.title("Monthly Revenue for Top 3 Countries")
plt.xticks(rotation=45)
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.show()
```
